```python
#na asociacne pravidla potrebujem mat v hlavicke nazvy produktov, v indexe order_id a hodnoty maju predstavovat qty
#Null hodnoty = 0
mybasket = mybasket.pivot(index ='ORDER_ID', columns ='PRODUCT_NAME_ID')['QTY_ORDERED'].fillna(0)
mybasket
```

```python
#kontrola dat
#mybasket.to_csv("data_ready.csv", index = True)
```

```python
#q_ty ordered >= 1 zakodovane 1, vsetko ostatne 0
def encoding_values(x):
    if x >= 1:
        return 1
    if x <= 0:
        return 0
    
my_basket_sets = mybasket.applymap(encoding_values)
```

```python
#my_basket_sets.to_csv("data_check.csv", index = True)
```

# Apriori algorithm

```python
#vyfiltrovanie najcastejsie sa vyskytujucich poloziek napriec nakupnymi kosikmi
frequent_itemsets = apriori(my_basket_sets, min_support=0.0035, use_colnames=True)

#pre support = 0.005 plati, ze produkt sa nachadzal aspon v 17-tich objednavkach (z 3587 obj.)
frequent_itemsets = frequent_itemsets.sort_values(by=['support'], ascending=False)

#pocet produktov v itemsets
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
my_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
```

```python
my_rules.sort_values(by=['support'], ascending=False)
```

# Results

```python
final_rules = my_rules[(my_rules['confidence'] >= 0.3) & (my_rules['lift'] >= 2)]
final_rules
```
